```python
from openai import OpenAI


API_KEY = "sk-...................A"
BASE_URL = "https://api.futurexailab.com"

client = OpenAI(
    api_key=API_KEY,
    base_url=BASE_URL
)

SYSTEM_PROMPT = """
You are an AI Study Buddy designed to help students learn.
You should teach concepts simply, ask follow-up questions, and provide quizzes on request.
"""


def ask_study_buddy(user_message):
    response = client.chat.completions.create(
        model="gemini-2.5-flash",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_message}
        ]
    )
    return response.choices[0].message.content
```
