```python
from openai import OpenAI
import os

API_KEY="sk..................."
BASE_URL="https://api.futurexailab.com"

def ask(prompt):
    client = OpenAI(
        api_key=API_KEY,
        base_url=BASE_URL
    )
    completion = client.chat.completions.create(
        model="gemini-2.5-flash",
        messages=[{"role": "user", "content": prompt}]
    )
    return completion.choices[0].message.content



# ZERO-SHOT PROMPTING

try:
    print("\n===== ZERO SHOT PROMPTING =====")
    prompt = "Explain machine learning in simple words."
    print(ask(prompt))
except Exception as e:
    print("An error occurred:", e)
```

```python
# ONE-SHOT PROMPTING

try:
    print("\n===== ONE SHOT PROMPTING =====")
    prompt = """
Example:
Q: What is supervised learning?
A: Supervised learning uses labelled data.

Now answer:
Q: What is unsupervised learning?
"""
    print(ask(prompt))
except Exception as e:
    print("An error occurred:", e)
```

```python
# FEW-SHOT PROMPTING

try:
    print("\n===== FEW SHOT PROMPTING =====")
    prompt = """
Example 1:
Input: What is regression?
Output: Regression predicts continuous values.

Example 2:
Input: What is classification?
Output: Classification predicts categories.

Now answer:
Input: What is clustering?
"""
    print(ask(prompt))
except Exception as e:
    print("An error occurred:", e)
```

```python
# CHAIN OF THOUGHT PROMPTING

try:
    print("\n===== CHAIN OF THOUGHT PROMPTING =====")
    prompt = """
Solve step by step:
Raju has 5 apples. He buys 7 more. He gives 3 to his friend.
How many apples does he have now?
"""
    print(ask(prompt))
except Exception as e:
    print("An error occurred:", e)
```

```python
# ROLE-BASED PROMPTING

try:
    print("\n===== ROLE BASED PROMPTING =====")
    prompt = """
You are a senior data science trainer.
Explain Random Forest in the simplest possible way.
"""
    print(ask(prompt))
except Exception as e:
    print("An error occurred:", e)
```

```python
# DOMAIN-SPECIFIC PROMPTING

try:
    print("\n===== DOMAIN SPECIFIC PROMPTING =====")
    prompt = """
You are a healthcare data scientist.
Explain how machine learning helps in predicting diabetes.
"""
    print(ask(prompt))
except Exception as e:
    print("An error occurred:", e)
```
