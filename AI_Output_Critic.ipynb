```python
from openai import OpenAI

# =======================
# CONFIG
# =======================
API_KEY = "sk...................Pio3A"
BASE_URL = "https://api.futurexailab.com"

client = OpenAI(
    api_key=API_KEY,
    base_url=BASE_URL
)

# =======================
# CRITIC FUNCTION
# =======================
CRITIC_PROMPT = """
You are an AI Output Critic.
Your job is to analyze the given content and produce:

1. Strengths
2. Weaknesses
3. Suggestions for improvement
4. Overall quality rating (1–10)

Keep the critique constructive, clear, and useful.
"""

def critique_output(user_text):
    messages = [
        {"role": "system", "content": CRITIC_PROMPT},
        {"role": "user", "content": user_text}
    ]

    try:
        response = client.chat.completions.create(
            model="gemini-2.5-flash",
            messages=messages
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {e}"
```
